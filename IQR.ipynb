```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
data=pd.read_csv("loan.csv")
data
```

```python

sns.boxplot(x="CoapplicantIncome",data=data)
plt.show()
```

```python
sns.displot(data["ApplicantIncome"])
```

```python
#outliers are not good for your data analysis
#so lets remove it using 2 methods  z score and iqr
```

```python
q1 = data["CoapplicantIncome"].quantile(0.25)
q3 = data["CoapplicantIncome"].quantile(0.75)
q1,q3
```

```python
IQR = q3 - q1
IQR
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range,max_range
```

```python
new_data = data[data["CoapplicantIncome"]<= max_range]
new_data
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_data)
```
