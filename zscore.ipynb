```python
import seaborn as sns
```

```python
data=pd.read_csv("loan.csv")
data
```

```python
sns.boxplot(x='CoapplicantIncome',data=data)
plt.show
```

```python
sns.displot(data["CoapplicantIncome"])
```

```python
minrange = data["CoapplicantIncome"].mean() - (3*data["CoapplicantIncome"].std()) # (mu - 3 sigma)
```

```python
maxrange = data["CoapplicantIncome"].mean() + (3*data["CoapplicantIncome"].std()) # (mu + 3 sigma)
```

```python
minrange,maxrange
```

```python
new_Data = data[data["CoapplicantIncome"]<=maxrange]
new_Data
```

```python
#zscore = x - mu / sigma 
z_score = (data["CoapplicantIncome"]-data["CoapplicantIncome"].mean())/data["CoapplicantIncome"].std()
```

```python
data["z_score"]= z_score
```

```python
data
```

```python
data[data["z_score"]<3]
```
